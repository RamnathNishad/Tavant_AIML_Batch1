```python
import pandas as pd
x=[3,4,5,6,7]

a=pd.Series(x)

print(a)
print(type(a))
```

# **Creating Series using othe objects like dictionary, tuple,list**

```python
dic={"name":["python","c","c++","java"],"popularity":[12,13,14,15],"rank":[1,4,3,2]}

var1 = pd.Series(dic)

print(var1)
```

```python
s=pd.Series(12,index=[1,2,3,4,5])
print(s)
print(type(s))
```

```python
# Works with missing data also
s1=pd.Series(12,index=[1,2,3,4,5,6,7])
s2=pd.Series(12,index=[1,2,3,4,5])

print(s1)
print()
print(s2)
print()

print(s1+s2)
```

# **Arithematic Operations**

```python
var =pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})

var["C"]=var["A"] + var["B"]
var["Python"]=var["A"]<=3
print(var)
```

# **Deleting and Inserting in Pandas**

```python
var =pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var.insert(1,"python",var["A"])
print(var)
```

```python
var =pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var.insert(1,"python",[11,12,13,14])
var["python"]=var["A"][1:]
print(var)
```

```python
#Delete
var1=var.pop("B")
var1
```

# **Handling missing Data in DataFrame**

```python
csv_1.dropna()   # drops all the rows and cols having NaN values

csv_1.dropna(axis=1)   # drops all the cols having missing values

csv_1.dropna(axis=0)   # drops all the rows having missing values

csv_1.dropna(how="all")  # delete rows if all cols NaN values else it will show them

csv_1.dropna(how="any")  # delete rows if any of the cols have NaN

csv_1.fillna('python')

csv_1.fillna(method="bfill",axis=0)  # fill data with next row/col or prev row/col data

csv_1.fillna({"a":"a_value","b":"b_value"})
```
