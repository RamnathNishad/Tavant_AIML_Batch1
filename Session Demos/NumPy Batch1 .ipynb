# **Copy vs View**

```python
import numpy as np

var = np.array([1,2,3,4])
cp=var.copy()

print("var:",var)
print("copy:",cp)
print(id(var))
print(id(cp))
var[1]=8
print("var:",var)
print("copy:",cp)
```

```python
x=np.array([9,8,7,6,5])
vi=x.view()
print("x:",x)
print("view:",vi)
print(id(x))
print(id(vi))
```

```python
x[1]=40
print("x:",x)
print("view:",vi)
```

```python
# Joining of Arrays: putting content of two or more arrays in a single array
var1=np.array([[1,2],[3,4]])
var2=np.array([[9,8],[7,6]])

ar_new = np.concatenate((var1,var2),axis=0)
print(var1)
print()
print(var2)
print()
print(ar_new)


```

```python
#Stack Function (Merging Arrays)
var1=np.array([1,2,3,4])
var2=np.array([9,8,7,6])

a_new=np.stack((var1,var2),axis=0)
print(a_new)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,8,7,6])

a_new=np.vstack((var1,var2))
print(a_new)
```

```python
var1=np.array([1,2,3,4])
var2=np.array([9,8,7,6])

a_new=np.hstack((var1,var2))
print(a_new)
```

```python
#Split arrays : breaks one array into multiple arrays
var= np.array([1,2,3,4,5,6])
print(var)

ar=np.array_split(var,3)
print()
print(ar)
print(ar[0])
```
