```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Iris.csv")
dataset
```

```python
# get the required data from the dataset
dataset.iloc[:,1:5]
```

```python
# extracted the input and output data from the dataset
x=dataset.iloc[:,1:5].values

y=dataset.iloc[:,-1].values

# we need to encode the labelled data in terms of numbers
from sklearn.preprocessing import LabelEncoder
labelencoder_y=LabelEncoder()
y=labelencoder_y.fit_transform(y)
```

```python
# divide the training and test data from the dataset x and y
# x  means it has both training and testing "INPUT" data
# y means it has both training and testing "OUTPUT" data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
# select the algorithm and train the model using that

from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)

```

```python
# predict the model output using test input
y_predic=logmodel.predict(x_test)
```

```python
y_predic
```

```python
y_test
```

```python
# find the error metric of the model prediction
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_predic)
```

```python
x=iris[['sepal_length']]
y=iris[['sepal_width']]
x.head()
```
