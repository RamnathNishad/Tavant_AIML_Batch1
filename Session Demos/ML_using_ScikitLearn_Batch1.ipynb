```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Iris.csv")
dataset
```

```python
# get the required data from the dataset
dataset.iloc[:,1:5]
```

```python
# extracted the input and output data from the dataset
x=dataset.iloc[:,1:5].values

y=dataset.iloc[:,-1].values

# we need to encode the labelled data in terms of numbers
from sklearn.preprocessing import LabelEncoder
labelencoder_y=LabelEncoder()
y=labelencoder_y.fit_transform(y)
```

```python
# divide the training and test data from the dataset x and y
# x  means it has both training and testing "INPUT" data
# y means it has both training and testing "OUTPUT" data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
# select the algorithm and train the model using that

from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)

```

```python
# predict the model output using test input
y_predic=logmodel.predict(x_test)
```

```python
y_predic
```

```python
y_test
```

```python
# find the error metric of the model prediction
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_predic)
```

```python
# using LinearRgresssion since we are depicting numerical width based on length
# model - 1
import seaborn as sns

iris=sns.load_dataset("iris")
iris.head()
```

```python
x=iris[['sepal_length']]
y=iris[['sepal_width']]
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

#print(x_train.head())
#print(x_test.head())
#print(y_train.head())
#print(y_test.head())


from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)

y_pred=lr.predict(x_test)

from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)

#print(y_pred[0:5])
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

```

```python
# importing the main dataset
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
df.head()


```

```python
df.info()
```

```python
# keeping only the required column
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]

# renaming column names
df.columns=['date','state','cured','deaths','confirmed']
df.head()
```

```python
df.tail()
```

```python
# some date to plot confirmed cases
today=df[df.date=='2020-07-17']
today
```

```python
#Sorting data w. r. t. number of confirmed cases
max_confirmed_cases =today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
# Getting states with maximum number of confirmed cases
top_states_confirmed =max_confirmed_cases[0:5]
top_states_confirmed
```

```python
# Making bar-plot for state with top confirmed cases
sns.barplot(x='state',y='confirmed',data=top_states_confirmed,hue='state')
plt.show()
```

```python
#Sorting data w. r. t. number of death cases
max_death_cases =today.sort_values(by='deaths',ascending=False)
max_death_cases

# Getting states with maximum number of death cases
top_states_deaths =max_death_cases[0:5]
top_states_deaths

# Making bar-plot for state with top deaths cases
sns.barplot(x='state',y='deaths',data=top_states_deaths,hue='state')
plt.show()
```
