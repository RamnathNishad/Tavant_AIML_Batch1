# **Line Chart**

```python
x=[1,2,3,4]
y=[2,4,6,8]

c=['r','g','b','y']

plt.bar(x,y,color=c)

plt.show()
```

```python
x=["python","c","c++","Java"]
y=[85,70,60,82]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)

c=['y','b','m','g']

plt.title("Tavant",fontsize=20)
plt.bar(x,y,width=0.4,color=c,edgecolor='r',linestyle=":",alpha=0.5,label="popularity")
plt.legend()
plt.show()
```

```python
x=["python","c","c++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)

c=['y','b','m','g']

width=0.2
plt.title("Tavant",fontsize=20)
plt.barh(x,y,width,color='r',label="popularity")
plt.barh(x,z,width,color='y',label="popularity1")

plt.xticks(rotation=50)
plt.legend()
plt.show()
```

# **Scatter Plot**

```python

def get_sizes(day):
  return [10,200,400,300,200,100,400]

day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]

colors=["r","y","g","b","r","g","r"]
sizes=[400,200,400,300,200,100,400]

plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g",cmap="Accent")
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)

plt.show()
```

# **Histogram Plot**

```python
no=[49,50,45,54,56,12,2,8,29,58]
plt.hist(no,color="b",bins=[10,20,30,40,50,60],edgecolor="r",orientation="horizontal")

plt.title("Tavant")
plt.xlabel("python")
plt.ylabel("No.")
plt.show()
```

# **Pie Graph**

```python
x=[10,20,30,40]
y=["c","c++","Java","python"]

plt.pie(x,labels=y,shadow=True,explode=[0.2,0.0,0.0,0.0],autopct="%0.1f%%",radius=1.3,startangle=180,labeldistance=1.1,textprops={"fontsize":15})

plt.title("Tavant Tech")
plt.legend(loc=2)

plt.show()
```

# **Seaborn**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


var = [1,2,3,4,5,6,7]
var_1= [2,3,4,1,5,6,7]

x_1=pd.DataFrame({"var":var, "var_1":var_1})

sns.lineplot(data=x_1,x='var',y='var_1')

plt.show()
```

# **Line Plot**

```python

y_1=sns.load_dataset("penguins")


sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex")

plt.show()
```

# **Scatter plot**

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",sizes=(100,60),alpha=1,markers=m,palette="Accent")

plt.show()
```

# **Headmap Plot**

```python
import numpy as np
var = np.linspace(1,10,8).reshape(2,4)
ar=np.array([["a0","a1","a2","a3"],
              ["b0","b1","b2","b3"]])

print(var)
v=sns.heatmap(var,annot=var,cmap="PuOr",linewidths=10,linecolor='y')
v.set_xlabel("python")
v.set_ylabel("Tavant")
sns.set(font_scale=1)

plt.show()
```

# **Count Plot**

```python
var = sns.load_dataset("tips")
print(var)

sns.countplot(data=var,x="sex",hue="smoker",palette='dark:y')
#sns.barplot(x="sex",y="size",data=var)
plt.show()
```

# **Pair Plot**

```python
var =sns.load_dataset("tips").head(20)

sns,sns.pairplot(var,hue="sex",hue_order=["Female","Male"],y_vars=["total_bill","tip"],kind="scatter",markers=["*","<"])  # scatter, kde, reg, hist
plt.show()
```

# **Cat Plot**

```python
var =sns.load_dataset("tips").head(20)
sns.barplot(x="day",y="total_bill",data=var)
sns.set_style("white")
plt.figure(figsize=(12,5))

#sns.catplot(x="size",hue="sex", data=var,kind="count")
plt.show()
```

# **Multiple Plots (Facet - Grid)**

```python
var =sns.load_dataset("tips")
fg = sns.FacetGrid(var,col="day",hue="sex",palette="summer")

fg.map(plt.bar,"total_bill","tip").add_legend()

plt.show()
```
